```python
class Bank:
    def __init__(self):
        self.client_details_list = []
        self.loggedin = False
        self.cash = 100
        self.TranferCash = False
        
    def register(self, name, ph, password):
        cash = self.cash
        conditions = True
        if len(str(ph)) > 10 or len(str(ph)) < 10:
            print("Invalid Phone number.. Please Enter a valide Phone number ")
            conditions = False
            
            
        if len(password) < 6 or len(password) > 8:
            print("Enter Password Greater then 6 digit and less then 8 digit")
            conditions = False 
            
        if conditions == True:
            print("Account created successfully..")
            self.client_details_list = [name , ph, password , cash]
            with open(f"{name}.txt", "w") as f:
                for details in self.client_details_list:
                    f.write(str(details)+ "\n")
                    
    def login(self, name , ph , password):
        with open(f"{name}.txt" , "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")
            if str(ph) in str(self.client_details_list):
                if str(password) in str(self.client_details_list):
                    self.loggedin = True
                    
            if self.loggedin == True:
                print(f"{name} logged in ")
                self.cash = int(self.client_details_list[3])
                self.name = name
                
            else:
                print("Wrong Information!! Please Provide correct info.. ")
                
    def add_cash(self, amount):
        if amount > 0:
            self.cash += amount
            with open(f"{name}.txt" , "r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")
            
            with open(f"{name}.txt" , "w") as f:
                f.write(details.replace(str(self.client_details_list[3]), str(self.cash)))
            print("Amount added successfully")
        
        else:
            print("Enter correct valur of amount")
            
    def Transfer_cash(self, amount , name , ph):
        with open(f"{name}.txt" , "r") as f:
            details = f.read()
            self.client_details_list = details.split("\n")
            if str(ph) in self.client_details_list:
                self.Transfer = True
                
          
        if self.Transfer == True:
            total_cash = int(self.client_details_list[3]) + amount
            sender_cash = self.cash - amount
            with open(f"{name}.txt" , "w") as f:
                f.write(details.replace(str(self.client_details_list[3]), str(total_cash)))
                
            with open(f"{self.name}.txt" , "r") as f:
                my_details = f.read()
                self.client_details_list = my_details.split("\n")
                
            with open(f"{self.name}.txt" , "w") as f:
                f.write(my_details.replace(str(self.client_details_list[3]), str(sender_cash)))
             
            print("Amount transfer successfullay to" , name,"-", ph )
            print("Your Balance = " , sender_cash)
            self.cash = sender_cash
            
    def password_change(self, password):
        if len(password) < 6 or len(password) > 8:
            print("Enter Password Greater then 6 digit and less then 8 digit")
        else:
            with open(f"{self.name}.txt" , "r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")
                
            with open(f"{self.name}.txt" , "w") as f:
                f.write(details.replace(str(self.client_details_list[2]), str(password)))
                
            print("New Password setup successfully...")  
            
    def ph_change(self , ph):
        if len(str(ph)) > 10 or len(str(ph)) < 10:
            print("Invalid Phone number ! please enter 10 digit number")
        else:
            with open(f"{self.name}.txt","r") as f:
                details = f.read()
                self.client_details_list = details.split("\n")

            with open(f"{self.name}.txt","w") as f:
                f.write(details.replace(str(self.client_details_list[1]),str(ph)))
            print("new Phone number set up successfully")
            
            
            
if __name__ == "__main__":
    Bank_Operate = Bank()
    print("WELCOME TO OUR PY BANK")
    print("1. Login")
    print("2. Create a bank account")
    user = int(input("Make your decision: "))
    
    if user == 1:
        print("For loggin in please enter required information")
        name = input("Enter your name: ")
        ph = int(input("Enter your phone number: "))
        password = input("Enter a password: ")
        Bank_Operate.login(name, ph, password)
        
        while True:
            if Bank_Operate.loggedin:
                print("1. Add amount")
                print("2. Check Balance")
                print("3. Transfer Balance")
                print("4. Edit profile")
                print("5. Logout")
                login_user = int(input("Make a decision: "))
                if login_user == 1:
                    print("Balance = ", Bank_Operate.cash)
                    amount = int(input("Enter Amount: "))
                    Bank_Operate.add_cash(amount)
                    print("\n1. back menu")
                    print("2. Logout")
                    choose = int(input("Make a decision: "))
                    if choose == 1:
                        continue
                        
                    elif choose == 2:
                        break
                            
                            
                elif login_user == 2:
                    print("Balacne =",Bank_Operate.cash)
                    print("\n1.back menu")
                    print("2.Logout")
                    choose = int(input())
                    if choose == 1:
                        continue
                    elif choose == 2:
                        break
                 
                elif login_user == 3:
                    print("Your Balance= ", Bank_Operate.cash)
                    amount = int(input("Enter amount: "))
                    if amount >= 0 and amount < Bank_Operate.cash:
                        name = input("Enter person name: ")
                        ph = input("Enter person phone number: ")
                        Bank_Operate.Transfer_cash(amount, name, ph)
                        print("\n1.back menu")
                        print("2.Logout")
                        choose = int(input())
                        if choose == 1:
                            continue
                        elif choose == 2:
                            break
                    
                    elif amount < 0 :
                        print("Enter please correct value of amount")

                    elif amount >  Bank_Operate.cash:
                        print("You do not have enough balance to transfer")   
                        
                elif login_user == 4:
                    print("1.Password change")
                    print("2.Phone Number change")
                    edit_profile = int(input())
                    if edit_profile == 1:
                        new_passwrod = input("Enter new Password: ")
                        Bank_Operate.password_change(new_passwrod)
                        print("\n1.back menu")
                        print("2.Logout")
                        choose = int(input())
                        if choose == 1:
                            continue
                        elif choose == 2:
                            break
                    elif edit_profile == 2:
                        new_ph = int(input("Enter new Phone Number: "))
                        Bank_Operate.ph_change(new_ph)
                        print("\n1.back menu")
                        print("2.Logout")
                        choose = int(input())
                        if choose == 1:
                            continue
                        elif choose == 2:
                            break

                elif login_user == 5:
                    break
                        
        
        
    if user == 2:
        print("Creating a new  Account")
        name = input("Enter Name: ")
        ph = int(input("Enter Phone Number: "))
        password = input("Enter password: ")
        Bank_Operate.register(name, ph, password)
          
```
